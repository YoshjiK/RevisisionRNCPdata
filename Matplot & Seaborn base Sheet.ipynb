````python
# ğŸ“Œ MATPLOTLIB & SEABORN CHEAT SHEET - DATA ANALYST RNCP NIVEAU 7 ğŸ“Š

---

# ğŸ¨ **IMPORTATION DES LIBRAIRIES**

```python
import matplotlib.pyplot as plt  # BibliothÃ¨que de base pour la crÃ©ation de graphiques
import seaborn as sns  # Extension de Matplotlib pour des graphiques plus esthÃ©tiques
import numpy as np  # BibliothÃ¨que pour le calcul numÃ©rique
import pandas as pd  # BibliothÃ¨que pour la manipulation des donnÃ©es
```

---

# ğŸ“Š **MATPLOTLIB : CrÃ©ation de Graphiques de Base**

### ğŸ”¹ **CrÃ©er une figure et un axe**
```python
fig, ax = plt.subplots(figsize=(8, 5))  # CrÃ©e une figure avec un axe de taille 8x5
```

### ğŸ”¹ **TracÃ© d'une courbe simple**
```python
x = np.linspace(0, 10, 100)  # GÃ©nÃ¨re 100 points entre 0 et 10
y = np.sin(x)  # Applique la fonction sinus
plt.plot(x, y, label='Sinus')  # Trace la courbe sinus
plt.xlabel('X')  # Nom de l'axe X
plt.ylabel('Y')  # Nom de l'axe Y
plt.title('Courbe Sinus')  # Titre du graphique
plt.legend()  # Ajoute une lÃ©gende
plt.show()  # Affiche le graphique
```

### ğŸ”¹ **Histogramme**
```python
data = np.random.randn(1000)  # GÃ©nÃ¨re 1000 valeurs alÃ©atoires suivant une distribution normale
plt.hist(data, bins=30, color='blue', edgecolor='black')  # CrÃ©e un histogramme avec 30 intervalles
plt.xlabel('Valeurs')  # Nom de l'axe X
plt.ylabel('FrÃ©quence')  # Nom de l'axe Y
plt.title('Histogramme des valeurs')  # Titre du graphique
plt.show()  # Affiche le graphique
```

### ğŸ”¹ **Scatter Plot (Nuage de points)**
```python
x = np.random.rand(50)  # 50 valeurs alÃ©atoires pour X
y = np.random.rand(50)  # 50 valeurs alÃ©atoires pour Y
plt.scatter(x, y, color='red', alpha=0.5)  # Trace un nuage de points avec transparence
plt.xlabel('X')  # Nom de l'axe X
plt.ylabel('Y')  # Nom de l'axe Y
plt.title('Nuage de points')  # Titre du graphique
plt.show()  # Affiche le graphique
```

---

# ğŸ¨ **SEABORN : Visualisation AvancÃ©e**

### ğŸ”¹ **Histogramme avec KDE (DensitÃ© de probabilitÃ©)**
```python
sns.histplot(data, bins=30, kde=True, color='purple')  # Affiche un histogramme avec courbe de densitÃ©
plt.title('Histogramme avec KDE')  # Titre du graphique
plt.show()  # Affiche le graphique
```

### ğŸ”¹ **Boxplot (BoÃ®te Ã  moustaches)**
```python
df = pd.DataFrame({'CatÃ©gorie': ['A', 'B', 'C', 'D'] * 10, 'Valeur': np.random.randn(40)})
sns.boxplot(x='CatÃ©gorie', y='Valeur', data=df, palette='Set3')  # Affiche un boxplot par catÃ©gorie
plt.title('Boxplot par catÃ©gorie')  # Titre du graphique
plt.show()  # Affiche le graphique
```

### ğŸ”¹ **Heatmap (Carte de chaleur)**
```python
corr_matrix = np.random.rand(5, 5)  # GÃ©nÃ¨re une matrice de corrÃ©lation alÃ©atoire
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)  # Affiche une carte de chaleur
plt.title('Heatmap de la matrice de corrÃ©lation')  # Titre du graphique
plt.show()  # Affiche le graphique
```

### ğŸ”¹ **Pairplot (Analyse multi-variÃ©e)**
```python
df = sns.load_dataset('iris')  # Charge le dataset iris
sns.pairplot(df, hue='species')  # Affiche les relations entre les variables par espÃ¨ce
plt.show()  # Affiche le graphique
```

---

# ğŸ¯ **PERSONNALISATION DES GRAPHIQUES**

### ğŸ”¹ **Modifier la taille de la figure**
```python
plt.figure(figsize=(10, 6))  # DÃ©finit une taille spÃ©cifique pour la figure
```

### ğŸ”¹ **Changer la couleur et le style des lignes**
```python
plt.plot(x, y, color='green', linestyle='--', linewidth=2)  # Personnalisation couleur et style
```

### ğŸ”¹ **Ajouter des annotations**
```python
plt.annotate('Point clÃ©', xy=(5, np.sin(5)), xytext=(7, 0.5),
             arrowprops=dict(facecolor='black', arrowstyle='->'))  # Ajoute une annotation avec une flÃ¨che
```

### ğŸ”¹ **Ajouter une grille**
```python
plt.grid(True, linestyle='--', alpha=0.7)  # Ajoute une grille avec style personnalisÃ©
```

---

# ğŸ“Œ **RÃ‰SUMÃ‰ DES GRAPHIQUES DISPONIBLES**

Voici tous les graphiques essentiels pour comprendre **Matplotlib et Seaborn** :

1ï¸âƒ£ **Courbe SinusoÃ¯dale** ğŸ“ˆ  (TracÃ© de courbe)
2ï¸âƒ£ **Histogramme** ğŸ“Š  (RÃ©partition des valeurs)
3ï¸âƒ£ **Nuage de points (Scatter Plot)** ğŸ”´  (Distribution entre deux variables)
4ï¸âƒ£ **Boxplot (BoÃ®te Ã  moustaches)** ğŸ“¦  (Analyse des valeurs extrÃªmes et distribution)
5ï¸âƒ£ **Heatmap (Carte de chaleur)** ğŸ”¥  (CorrÃ©lations entre variables)
6ï¸âƒ£ **Pairplot (Analyse multi-variÃ©e)** ğŸ“Œ  (Relations entre plusieurs variables)
7ï¸âƒ£ **Personnalisation des graphiques** ğŸ¨  (Couleur, styles, annotations)

Tous ces graphiques sont personnalisables avec :
- ğŸ“ **Taille** (`figsize=(10, 6)`)  (Dimension de la figure)
- ğŸ¨ **Couleurs** (`color='green'`)  (Choix des couleurs des Ã©lÃ©ments du graphique)
- ğŸ¹ **Annotations** (`plt.annotate(...)`)  (Ajout d'annotations sur un graphique)
- ğŸ“Š **LÃ©gendes** (`plt.legend()`)  (Ajout d'une lÃ©gende pour identifier les Ã©lÃ©ments du graphique)

ğŸ’¡ **Astuce :** Utilise **Seaborn** pour des visualisations plus avancÃ©es et esthÃ©tiques, et **Matplotlib** pour plus de contrÃ´le sur les graphiques ! ğŸš€
````
